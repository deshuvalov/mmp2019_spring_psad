```python
from scipy.stats import linregress

def draw_linear(x, y, x_name='', y_name=''):
    "Scatter x-y points and draw a LR line."
    slope, intercept, r, _, _ = linregress(x, y)
    x_lims = np.array([x.min() - 1, x.max() + 1])
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.plot(x, y, 'g+', label='Objects')
    plt.plot(x_lims, intercept + slope * x_lims, 'r--', label=r"LR line with $\rho \approx {:.3f}$".format(r))
    plt.legend()
    plt.show()
```

```python
from sklearn.metrics import confusion_matrix

def draw_cm(x_bin, y_bin, x_name='', y_name=''):
    "Draw a confusion matrix for binary 1-D arrays x_bin and y_bin."
    cm = confusion_matrix(x_bin, y_bin)
    plt.matshow(cm)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.colorbar()
    plt.grid(False)
    plt.show()
```

```python
from itertools import combinations
from collections import defaultdict

def apply_pair_statistics(S, pair_statistics, statistics_names):
    """Apply pair statistics to each pair in a dataset S 
    and return a DataFrame containing the results.
    
    Parameters
    ----------
    S: array of shape (N, D)
        A dataset numpy array.
    pair_statistics: list of callables
        A list of statistics -- functions requiring two array arguments (x and y) 
        and returning a number (statistic's value)
    statistics_names: list of strings
        A list of statistics names.
    """
    res = defaultdict(list)

    for i, j in combinations(range(S.shape[1]), 2):
        x_name, y_name = f'X{i+1}', f'X{j+1}'
        x, y = S.T[i], S.T[j]
        res["Pair"].append(x_name + ' - ' + y_name)
        for statistic, name in zip(pair_statistics, statistics_names):
            res[name].append(statistic(x, y))

    res = pd.DataFrame(res)
    return res
```

Вычислите для каждой пары переменных коэффициент корреляции Пирсона.

```python
def pearson(x, y):
    "Calculate a Pearson correlation coefficient given 1-D data arrays x and y."
    x_mean, y_mean = x.mean(), y.mean()
    return np.mean((x - x_mean) * (y - y_mean)) / np.std(x) / np.std(y)

res_pearson = apply_pair_statistics(S, [pearson], ['Pearson'])
res_pearson
```

```python
def pearson_test(x, y):
    return stats.pearsonr(x, y)[0]

apply_pair_statistics(S, [pearson, pearson_test], ['Pearson', 'Pearson_test'])
```

Сопоставьте каждой из переменных  $X_1,\dots,X_4$  соответствующую порядковую переменную.

```python
def rank(x, method='average'):
    "Assign ranks to data and return the resulting ranking array."
    _, indices, counts = np.unique(x, return_counts=True, return_inverse=True)
    
    ranks = np.ones_like(counts)
    if method == 'average':
        ranks[1:] = np.cumsum(counts)[:-1] + 1
        ranks = ranks + (counts - 1) / 2
    elif method == 'min':
        ranks[1:] = np.cumsum(counts)[:-1] + 1
    elif method == 'max':
        ranks = np.cumsum(counts)
    return ranks[indices]
    
S_r = np.array([rank(x) for x in S.T]).T
```

Вычислите для каждой из пар переменных коэффициенты корреляции Спирмена и Кендалла ($\tau$-коэффициент).

```python
def spearman(x_r, y_r):
    "Calculate a Spearman correlation coefficient given 1-D ranking arrays x_r and y_r."
    return pearson(x_r, y_r)

def kendall(x_r, y_r):
    "Calculate a Kendall tau correlation coefficient given 1-D ranking arrays x_r and y_r."
    xless = x_r[:, None] < x_r[None, :]
    yless = y_r[:, None] < y_r[None, :]
    ygreater = y_r[:, None] > y_r[None, :]
    
    n_c = (xless * yless).sum()
    n_d = (xless * ygreater).sum()
    
    n = x_r.shape[0]
    n_0 = n * (n - 1.0) / 2.0
    
    _, t_1 = np.unique(x_r, return_counts=True)
    n_1 = np.sum(t_1 * (t_1 - 1.0) / 2.0)
    
    _, t_2 = np.unique(y_r, return_counts=True)
    n_2 = np.sum(t_2 * (t_2 - 1.0) / 2.0)
    
    return (n_c - n_d) / np.sqrt((n_0 - n_1) * (n_0 - n_2))

res_ranked = apply_pair_statistics(S_r, [spearman, kendall], ['Spearman', 'Kendall'])
res_ranked
```

```python
def spearman_test(x_r, y_r):
    return stats.spearmanr(x_r, y_r)[0]

def kendall_test(x_r, y_r):
    return stats.kendalltau(x_r, y_r)[0]

apply_pair_statistics(S_r, [spearman, spearman_test, kendall, kendall_test], ['Spearman', 'Spearman_test', 'Kendall', 'Kendall_test'])
```

```python
res_ranked_min = apply_pair_statistics(np.array([rank(x, 'min') for x in S.T]).T, [spearman, kendall], ['Spearman, min rank', 'Kendall, min rank'])
res_ranked_min
```

```python
res_ranked_max = apply_pair_statistics(np.array([rank(x, 'max') for x in S.T]).T, [spearman, kendall], ['Spearman, max rank', 'Kendall, max rank'])
res_ranked_max
```

Вычислите для каждой пары бинаризованных переменных коэффициенты близости Жаккарда и $\phi$.

```python
def jaccard(x_bin, y_bin):
    "Compute the Jaccard coefficient given 1-D boolean arrays x_bin and y_bin."
    return np.inner(x_bin, y_bin) / np.maximum(x_bin, y_bin).sum()

def phi(x_bin, y_bin):
    "Compute the phi coefficient given 1-D boolean arrays x_bin and y_bin."
    return pearson(x_bin, y_bin)

res_bin = apply_pair_statistics(S_bin, [jaccard, phi], ['Jaccard', 'Phi'])
res_bin
```

```python
from sklearn.metrics import jaccard_score, matthews_corrcoef
```

```python
def jaccard_test(x_bin, y_bin):
    "Compute the Jaccard coefficient given 1-D boolean arrays x_bin and y_bin."
    return jaccard_score(x_bin, y_bin)

def phi_test(x_bin, y_bin):
    "Compute the phi coefficient given 1-D boolean arrays x_bin and y_bin."
    return matthews_corrcoef(x_bin, y_bin)

apply_pair_statistics(S_bin, [jaccard, jaccard_test, phi, phi_test], ['Jaccard', 'Jaccard_test', 'Phi', 'Phi_test'])
```

### Нормальность

2. Проверим нормальность двух выборок (для каждой из диет). Для этого нарисуйте qq-график и отметьте на нем прямую линию, соответствующую идеальным теоретическим квантилям нормального распределения. Похожи ли данные на нормальное распределение?

```python
rats
```

```python
def qqplot(data, name):
    quantiles = np.linspace(1e-6, 1.0 - 1e-6, len(data))
    normal_q = stats.norm.ppf(quantiles)
    data_q = np.quantile((data - data.mean()) / data.std(ddof=1), quantiles)
    
    plt.figure()
    plt.plot(normal_q, data_q, ls='', marker='o', label='Restricted diet')
    plt.xlabel('Normal quantiles')
    plt.ylabel('Sample quantiles')
    plt.title(f'{name} diet')
    plt.legend()
    plt.grid(True)

    m, b = data_q.std(ddof=1), data_q.mean()
    ref_range = np.asarray([normal_q.min(), normal_q.max()])
    ref_line = ref_range * m + b

    plt.plot(ref_range, ref_line, '--', linewidth=2)
```

```python
qqplot(x1, 'restricted')
qqplot(x2, 'ad libitum')
```

**Ответ:** Я бы не сказал, что данные нормальные. Есть перекос в сторону меньших данных.

### Сравнение средних

Проверьте нулевую гипотезу

$H_0$: продолжительность жизни крыс не меняется при ограничении диеты.

против следующих альтернатив:

$H_1$: крысы на строгой диете живут дольше.

$H_1$: средняя продолжительность жизни крыс меняется при ограничении диеты. 

Проверить правильность вычислений вы можете с помощью библиотечной функции `scipy.stats.ttest_ind`, которая реализует двусторонний критерий Стьюдента для двух независимых выборок.

```python
n1, n2 = x1.shape[0], x2.shape[0]
x1_mean, x2_mean = x1.mean(), x2.mean()
S1_2, S2_2 = x1.var(ddof=1), x2.var(ddof=1)

T = (x1_mean - x2_mean) / np.sqrt(S1_2 / n1 + S2_2 / n2)
nu = (S1_2 / n1 + S2_2 / n2) ** 2 / (S1_2 ** 2 / (n1 ** 3 - n1 ** 2) + S2_2 ** 2 / (n2 ** 3 - n2 ** 2))

print(T, nu)
```

```python
n1, n2
```

```python
S1_2, S2_2
```

```python
stats.t.cdf(np.abs(T), df=nu), 1 - stats.t.cdf(np.abs(T), df=nu)
```

```python
stats.ttest_ind(x1, x2, equal_var=False)
```

**Ответ:** Поскольку уровень значимости крайне мал, мы отвергаем нулевую гипотезу при тестировании как против односторонней, так и двусторонней гипотезы.
